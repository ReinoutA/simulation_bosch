```python
import matplotlib.pyplot as plt

# Convert lists to a dictionary
total_transition_time_map = {'FCFS': 50, 'CR': 80, 'SDD': 30}

# Unpack keys and values for plotting
labels, total_transition_time = zip(*total_transition_time_map.items())

plt.figure(figsize=(3, 2))
plt.bar(labels, total_transition_time, color=['tab:blue', 'tab:green', 'tab:red'])
plt.show()
```
